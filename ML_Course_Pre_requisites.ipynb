```python
# Q1
import numpy as np
import re

V1 = np.random.random(100)
V1.sort()

print(V1)
print("\n")

# Q2

print(V1 * 3)
print("\n")

# Q3

mean = np.mean(V1)
sd = np.std(V1)

print("Mean: " + str(mean))
print("Standard Deviation: " + str(sd))
print("\n")

# Q4

matrix = np.random.random((4, 3))
print(matrix)
print("\n")

matrix_flattened = matrix.flatten()
print(matrix_flattened)
print("\n")

# Q5

S1 = "I am a great learner. I am going to have an awesome life."
print(S1)
print("\n")
print("Occurrence count of 'am' in S1: " + str(S1.lower().count("am")))
print("\n")

# Q6

S2 = "I work hard and shall be rewarded well."
S3 = S1 + S2
print(S3)
print("\n")

# Q7

words_arr = re.split(r'[.\s]+', S3)
print("Length of words_arr: " + str(len(words_arr)))

# Q8

words_to_remove = ["i", "am", "to", "and"]

words_arr_2 = [word for word in words_arr if word.lower() not in words_to_remove and len(word) <= 6]
print(words_arr_2)
print("Length: " + str(len(words_arr_2)))
print("\n")

# Q9

date = "01-JUN-2021"
month_map = {
    "JAN": 1,
    "FEB": 2,
    "MAR": 3,
    "APR": 4,
    "MAY": 5,
    "JUN": 6,
    "JUL": 7,
    "AUG": 8,
    "SEP": 9,
    "OCT": 10,
    "NOV": 11,
    "DEC": 12
}

date_arr = date.split("-")
print(date)
print("Month: " + str(month_map[date_arr[1]]))
print("\n")

# Q10
import pandas as pd

df = pd.read_csv("/content/table_data.csv")
df["Location"] = df["City"] + ", " + df["State"]
print(df)
print("\n")
df.to_excel("output.xlsx")

# Q11
import matplotlib.pyplot as plt

plt.plot(V1, color="red")

# Q12

V2 = V1 * V1

plt.plot(V2)
```
